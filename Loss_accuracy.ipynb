### Plot loss and accuracy

```python
%matplotlib inline

#coding: utf-8
import matplotlib.pyplot as plt
import numpy as np
import os
import json
plt.style.use('ggplot') 

import predict

# 
def plot_graph(result_dir = 'result'):
    f = open(os.path.join(result_dir, 'log'), 'r')
    log = json.load(f)
    iter = [d["iteration"] for d in log]
    epoch = [d["epoch"] for d in log]
    loss_train = [d["main/loss"] for d in log]
    loss_test  = [d["validation/main/loss"] for d in log]
    rec_train = [d["main/recall"] for d in log]
    rec_test  = [d["validation/main/recall"] for d in log]
    pre_train = [d["main/precision"] for d in log]
    pre_test  = [d["validation/main/precision"] for d in log]
    
    plt.scatter(iter, loss_train, label="train")
    plt.scatter(iter, loss_test , label="test")
    plt.title('Loss')
    plt.ylim([0.0, 0.005])
    plt.legend()
    plt.show()

    plt.scatter(iter, rec_train, label="train")
    plt.scatter(iter, rec_test , label="test")
    plt.title('Recall')
    plt.legend()
    plt.show()

    plt.scatter(iter, pre_train, label="train")
    plt.scatter(iter, pre_test , label="test")
    plt.title('Precision')
    plt.legend()
    plt.show()

    
plot_graph(result_dir='result_0520/')
    
```
